```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
Titanic_Disaster = pd.read_csv('train.csv')
Titanic_Disaster.head()
```

```python
sns.displot(data=Titanic_Disaster, x="Age",kind="kde");
```

```python
sns.displot(data=Titanic_Disaster,x='Fare',kind='kde');
```

```python
svl_rate = Titanic_Disaster.groupby('Embarked')['Survived'].value_counts(normalize = True)
svl_rate
svl_rate.astype('float64')
```

```python
svl_rate = Titanic_Disaster.groupby('Pclass')['Survived'].value_counts(normalize = True)
svl_rate
svl_rate.astype('float64')
```

```python
print('Percentage of 1st Passenger Class Survivors     : ', svl_rate[1] * 100, '%')
print('Percentage of 1st Passenger Class Non-Survivors : ', svl_rate[2] * 100, '%')
print('Percentage of 2nd Passenger Class Survivors     : ', svl_rate[1] * 100, '%')
print('Percentage of 2nd Passenger Class Non-Survivors : ', svl_rate[2] * 100, '%') 
print('Percentage of 3rd Passenger Class Survivors     : ', svl_rate[1] * 100, '%')
print('Percentage of 3rd Passenger Class Non-Survivors : ', svl_rate[3] * 100,'%')
```
